>#### Functional Plot

```python
x1 = np.linspace(0,5,20)
y1 = x1**2
y2 = x1**3
y3 = x1**4
y4 = x1**5
```

```python
# plots y = x^2 graph
plt.plot(x1,y1)
plt.title("f(x) = x^2")
plt.xlabel("x")
plt.ylabel("f(x)")
```

>#### Multiple plots

```python
# define (total_rows, total_cols, plot number)
plt.subplot(2, 2, 1)
plt.plot(x1,y1)
plt.title("f(x) = x^2")
plt.xlabel("x")
plt.ylabel("f(x)")

plt.subplot(2, 2, 2)
plt.plot(x1,y2,'r')
plt.title("f(x) = x^3")
plt.xlabel("x")
plt.ylabel("f(x)")

plt.subplot(2, 2, 3)
plt.plot(x1,y3,'b')
plt.title("f(x) = x^4")
plt.xlabel("x")
plt.ylabel("f(x)")

plt.subplot(2, 2, 4)
plt.plot(x1,y4,'g')
plt.title("f(x) = x^5")
plt.xlabel("x")
plt.ylabel("f(x)")
```

>#### Subplots


```python
# generate a framework for 3 graphs in a single row

fig_1, axes_1 = plt.subplots(figsize=(8,4), nrows=1, ncols=3)
plt.tight_layout()

# access plots by index
axes_1[1].set_title('Plot 2')
axes_1[1].set_xlabel('x')
axes_1[1].set_ylabel('f(x)')
axes_1[1].plot(x1,y1,color='red')
```

>#### Using Figure Objects

Figure is an object that contains all plot elements, axes, and other chart attributes

```python
# generate plots inside of plots

# figsize=(width,height)
fig_2 = plt.figure(figsize=(5,4))

#.add_axes([x_coord,y_coord, x_axis_size, y_axis_size])
axes_2 = fig_2.add_axes([0,0,1.2,1.2])
axes_2.set_xlabel("x")
axes_2.set_ylabel("f(x)")
axes_2.set_title("f(x) = x^2 and f(x) = x^3")
axes_2.plot(x1,y1,label='x^2')
axes_2.plot(x1,y2,label='x^3')
axes_2.legend(loc=0)

axes_3 = fig_2.add_axes([0.3,0.45,0.5,0.5])
axes_3.set_xlabel("x")
axes_3.set_ylabel("f(x)")
axes_3.set_title("f(x) = x^4 and f(x) = x^5")
axes_3.plot(x1,y3, color='green',label='x^4')
axes_3.plot(x1,y4, color='red' , label='x^5')
axes_3.legend(loc=0)
```

>#### Appearance Options

`axes.plot() arguments...`
- `color = <color_name>`
- `alpha = <value>`: transparency
- `lw = <num>`: number of times thicker
- `ls = '-'`: dots/solid lines
- `marker='x'`: puts marker at x,y defined points
- `markersize = <num>`: 
- `markercolor = 'y'`
- `markeredgecolor = 'y'`
- `markeredgewidth = <num>`

Refs:
- Full list of colors: https://en.wikipedia.org/wiki/Web_colors
- Linestyles:
    - `'-'` = solid
    - `':'` = dotted
    - `'--'` = dashed
    - `'-.'` = dashdot
- Markers:
    - `'.'` = point
    - `'o'` = circle
    - `'s'` = square
    - `'+'` = plus
    - `'x'` = cross
    - `'D'` = diamond


```python
fig_3 = plt.figure(figsize=(8,6))
axes_3 = fig_3.add_axes([0,0,1,1])
axes_3.plot(x1,y1,
        color = 'green',
        alpha = 0.8,
        lw = 2,
        ls = '-',
        marker='o',
        markersize = 5,
        markerfacecolor = 'white',
        markeredgecolor = 'red',
        markeredgewidth = 2)

# add a grey grid
axes_3.grid(True, color='0.3')
# add a chart color
axes_3.set_facecolor('LightSteelBlue')
```

>#### Working with Pandas Dataframe

```python
# dataframe created to show the sales of icecream at different day temperatures in a month
temp_arr = np.random.randint(30,50,30)
sales_arr = np.random.randint(300,600,30)

ice_df = pd.DataFrame({
    'Temperature': temp_arr,
    'Sales': sales_arr
})

# creating x and y variables based on cols
x_temp = np.array(ice_df.iloc[:,0])
y_sales = np.array(ice_df.iloc[:,1])
```

>#### Bar Charts

```python
# average scores of students in a year
names = ['A','B','C','D','E','F','G','H','I','J']
scores = np.random.randint(30,100,10)
variance = np.random.randint(1,10,10) # deviation in student scores
```

```python
# plotting number of boys and girls in different divisions of a class
class8_boys = np.random.randint(20,60,5)
class8_girls = np.random.randint(20,40,5)
space = np.arange(5) # just a placeholder for plotting
```

```python
# bar plot 2: side by side
plt.bar(space, class8_boys, width=0.45, label='Boys', edgecolor='k', color='green')
plt.bar(space+0.45, class8_girls, width=0.45, label='Girls', edgecolor='k')
plt.xticks(space+0.45/2, ('A','B','C','D','E'))
plt.legend(loc='upper center')
```

```python
# bar plot 2: stacked
divs = ['A','B','C','D','E']

plt.bar(divs, class8_boys, width=0.45, label='Boys', edgecolor='k')
plt.bar(divs, class8_girls, width=0.45, label='Girls', bottom=class8_boys,edgecolor='k')
plt.legend(loc=1)
```

```python
#bar plot 2: stacked percentage

class8_boys_percent = [(boy/(boy+girl))*100 for boy,girl in zip(class8_boys, class8_girls)]
class8_girls_percent = [(girl/(boy+girl))*100 for boy,girl in zip(class8_boys, class8_girls)]
plt.bar(divs, class8_boys_percent, width=0.45, label='Boys', edgecolor='k')
plt.bar(divs, class8_girls_percent, width=0.45, label='Girls', bottom=class8_boys_percent,edgecolor='k')
plt.legend(loc=1)
```
